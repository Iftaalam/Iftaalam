```python
dataset.columns= ['Gender','Age' ,'Marital_Status','Living_with_family','Current_living_place', 'Kn1','Kn2','Kn3','Kn4','Att1','Att2','Att3','Prac1','Prac2','Prac3','Prac4']
```

```python
#In the code above .columns is used to call the column in the dataset 
#Here dataset is the name of variable which I have declared above 
#dataset variable represents the excel file on which we are working
dataset
```

```python
dataset.replace({"Male(পুরুষ)":'Male',"Female(মহিলা)":'Female',"18-25(১৮-২৫)":'18-25',"26-35 (২৬-৩৫)":'26-35',"36-45 (৩৬-৪৫)":'36-45',"46-55 (৪৬-৫৫)":'46-55',"Unmarried (অবিবাহিত)":'Unmarried',"Married (বিবাহিত)":'Married',"Other (Separated, Divorced, Widowed){অন্যান্য (পৃথক, বিবাহবিচ্ছেদ, বিধবা)}":'Other',"Yes(হ্যাঁ)":'Yes',"No(না)":'No',"Maybe (হতে পারে)":'Maybe',"Outside Dhaka (ঢাকার বাইরে)":'Outside Dhaka',"Dhaka (ঢাকা)":'Dhaka',"Agree (একমত)":'Agree',"Neutral (নিরপেক্ষ)":'Neutral',"Disagree (অসম্মত)":'Disagree',"Yes (হ্যাঁ)":'Yes',"No (না)":'No' },inplace =True)
```

```python
dataset
```

```python
#Here we are again declaring a new variable.
#We are creating a new excel file with the processed data
#Here Updated_processed_raw_data is the name of my new excel file which will be saved in jupiter notebook.
#.xlsx is the format of newly created the excel file
excel_file = pd.ExcelWriter("Updated_processed_raw_data.xlsx")
```

```python
#It is a conversion code
dataset.to_excel(excel_file)
```

```python
#This code will save the new created excel file with the  processed data
excel_file.save()
```
